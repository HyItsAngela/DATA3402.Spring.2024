````python
# Write you solution here

def make_b(n, m):

  matrix = [[empty] * m for i in range(n)]

  for i in range(n): # outer loop iterates over EACH row of the matrix. "i" = current row index. Will run "n" times. "j" = current column index
    print(" ---" * m)
    for j in range(m): # inner loop iterates over each ELEMENT(cell) in the current row of the matrix.
      print(f"| {matrix[i][j]} ", end = "") # prints each element within the vertical lines of the board.
                                            # {matrix[i][j]} gets the value of the element at the current row "i" and column "j" from the matrix.
                                            # end parameter prints without a new line (to make it horizontal)
    print("|")

  print(" ---" * m)
````

*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has won, respectively. Here are some example inputs you can use to test your code:

````python
# Write you solution here

def game_status(matrix):

  n = len(matrix)

# check rows, cols, and diag for win
  for i in range(n): # iterates over each row
    if all(matrix[i][j] == matrix[i][0] and matrix[i][0] !=0 for j in range(n)):
      return matrix[i][0] # player ? wins in a row
    if all(matrix[j][i] == matrix[0][i] and matrix[0][i] !=0 for j in range(n)):
      return matrix[0][i] # player ? wins in a col

  if all(matrix[i][i] == matrix[0][0] and matrix[0][0] != 0 for i in range(n)):
    return matrix[0][0] # player ? wins in a main (left) diag
  if all(matrix[i][n-i-1] == matrix[0][n-1] and matrix[0][n-1] != 0 for i in range(n)):
    return matrix[0][n-1] # player ? wins in a right diag

  # check if game is incomplete or draw
  for i in range(n):
    for j in range(n):
      if matrix[i][j] == 0:
        return -1 # incomplete
  return 0 # draw
````

````python
# Write you solution here

empty = 0
#establish players
player_1 = 1
player_2 = 2

# player pieces
player_1_piece="X"
player_2_piece="O"
empty_space=" "

space_character= { player_1: player_1_piece,
                   player_2: player_2_piece,
                   empty: empty_space }

def print_b(board):
  for i in board: # "i" = row in board
    print(" ---" * len(i)) # len determines the length (how many cells are in the row)/amount of sets of dashes needed
    for j in i: # "j" represents the cell in "i"(row)
      print(f"| {space_character[j]} ", end = "")
    print("|")
  print(" ---" * len(i))

def make_b(n, m):
  board = [[empty] * m for i in range(n)]
  print_b(board)
  return board

def place_piece(board, player, x, y, verbose = True):

  # check is move exist on board
  if 0 <= x < len(board) and 0 <= y < len(board[0]):

    # check if cell is empty
    if board[x][y] == empty:

      # place player's mark in cell
      board[x][y] = player
      print_b(board)
      return True # successful move
    else:
      print("Cell is already occupied.")
  else:
    print("Move is out of bounds.")
  return False # unsuccessful move
````

````python
# Write your solution here

def print_message(message,verbose=True):
    if verbose:
        print(message)

def parse_mark(board, l_string):

  # find the board dimensions
  n = len(board)
  m = len(board[0]) if board else 0

  # row names
  row_names=list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
  row_map=dict(zip(row_names,range(n)))

  # col names
  column_names = list(map(str, range(1, m + 1)))
  column_map = dict(zip(column_names, range(m)))
  print("   ", end="")

  # converts input 'l-string' from an integer into a string
  if isinstance(l_string, int):
    l_string = str(l_string)
    
  if not isinstance(l_string,str):
    print_message("Bad Input. Location must be string.")
    return False

  if len(l_string)!=2:
    print_message("Bad Input. Location must be 2 characters.")
    return False

  row=l_string[0].upper()
  col=l_string[1].upper()

  if not row in row_names:
    print_message(f"Bad Row: {row}")
    return False

  if not col in column_names:
    print_message(f"Bad Column: {col}")
    return False

  return row_map[row],column_map[col]


def place_parse_piece(board, player, l_string, verbose=True):
    # find the board dimensions
    n = len(board)
    m = len(board[0]) if board else 0

    # row names
    row_names = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    row_map = dict(zip(row_names, range(n)))

    # col names
    column_names = list(map(str, range(1, m + 1)))
    column_map = dict(zip(column_names, range(m)))

    print("   ", end="")

    # converts input 'l-string' from an integer into a string
    if isinstance(l_string, int):
        l_string = str(l_string)

    if not isinstance(l_string, str):
        print_message("Bad Input. Location must be string.")
        return False

    if len(l_string) != 2:
        print_message("Bad Input. Location must be 2 characters.")
        return False

    row = l_string[0].upper()
    col = l_string[1].upper()

    if row not in row_names:
        print_message(f"Bad Row: {row}")
        return False

    if col not in column_names:
        print_message(f"Bad Column: {col}")
        return False

    x, y = row_map[row], column_map[col]

    # check if move exists on board
    if 0 <= x < len(board) and 0 <= y < len(board[0]):

        # check if cell is empty
        if board[x][y] == empty:

            # place player's mark in cell
            board[x][y] = player
            if verbose:
                print(f"Player {player} marked cell ({row}{col}).")
            print_b(board)  # print modified board
            return True  # successful move
        else:
            print("Cell is already occupied.")
    else:
        print("Move is out of bounds.")
    return False  # unsuccessful move
````

````python
# Write your solution here
import copy

def generate_moves(board,player,current_player,depth, keep_all_moves=False):
    if depth==0:
        return list()

    moves=list()

    # Look through the board for pieces belonging to a specific player
    for i in range(len(board)):
        for j in range(len(board[0])):
            if board[i][j]==empty:
                # Create a new copy of the board
                new_board=copy.deepcopy(board)
                # place player's mark in a empty cell
                new_board[i][j] = current_player

                # If so, generate subsequent moves
                next_moves = generate_moves(new_board,player,
                                                switch_player(current_player),
                                                depth-1,
                                                keep_all_moves=keep_all_moves)
                moves.append((i, j))
    return moves
````

````python
def computergic(board, player, depth): # computer logic, function explains thought process behind computers moves - Algorithm
  if depth == 0: # recursion will stop/amount of moves ahead calculated
    return None

  # 1st option (block): check, can the opponent win in the next move? If so, block them.
  opp_player = switch_player(player)
  oppwins = generate_moves(board, opp_player, opp_player, 1, keep_all_moves = True)
  for move in oppwins:
    new_board = copy.deepcopy(board)
    new_board[move[0]][move[1]] = opp_player
    if game_status(new_board) == opp_player:  # If blocking this move prevents opponent from winning
        return move  # Block the opponent's winning move

  # 2nd option (win): check, can the computer win in the next move?
  compwins = generate_moves(board, player, player, 1, keep_all_moves = True)
  if compwins:
    return compwins[0] # if a move is found, return the winning move
````

````python
def place_piece_com(board, player, row, col):
  board[row][col] = player

def computer(board, player):
  move = computergic(board, player, depth = 2)
  if move is not None:
    row, col = move
    place_piece_com(board, player, row, col)
  else:
    print("computer could not make move.")
````
