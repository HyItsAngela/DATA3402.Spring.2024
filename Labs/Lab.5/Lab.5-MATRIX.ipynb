2. Add the following methods:
    * `shape()`: returns a tuple `(n,m)` of the shape of the matrix.
    * `transpose()`: returns a new matrix instance which is the transpose of the matrix.
    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.
    * `to_list()`: which returns the matrix as a list of lists.
    *  `block(n_0,n_1,m_0,m_1)` that returns a smaller matrix located at the n_0 to n_1 columns and m_0 to m_1 rows. 
    * (Extra credit) Modify `__getitem__` implemented above to support slicing.
        

```python
class Matrix:
    def __init__(self,*args):
        if len(args)==2:
            #given two arguments, n and m. Initialize matrix with zeroes of size (nxm)
            self.rows=args[0] #m
            self.col=args[1] #n
            self.size=[[0]*self.col for _ in range(self.rows)] #zero_matrix
        elif len(args)==1 and isinstance(args[0],list): #(M,list)
            if self.is_matrix(args[0]): #is_matrix(args[0]=(M))
                self.size=[row[:] for row in args[0]] #create new list of lists with same values as args[0]. row[:]=copy of list.
                self.rows=len(args[0])
                self.col=len(args[0][0])
            else:
                return "Error 2: Improper matrix size"
        else:
            return "Error 1: Improper value"
            
    #EXTRA CREDIT: modified __getitem__ to handle splices
    def __getitem__(self,key):
        if isinstance(key,slice):
            return [row[key] for row in self.size]
        else:
            return self.size[key]
    
    def is_matrix(self,M): #M=args[0] essentailly, but args[0]=list of list represention of matrix.
        if isinstance(M,list): #M=instance of Matrix
            row_length=len(M[0]) if M else 0
            return all(len(row)==row_length for row in M)
        return False
    
    def __str__(self):
        return str(self.size)
    
    #new code
    def shape(self):
        return (self.rows,self.col)
    
    #new code
    def transpose(self):
        return Matrix([list(row) for row in zip(*self.size)])
    
    #new code
    def row(self,n):
        return Matrix([self.size[n]])
    
    #new code
    def column(self,n):
        return Matrix([[row[n]] for row in self.size])
    
    #new code
    def to_list(self):
        return [row[:] for row in self.size]
    
    #new code
    def block(self,n_0,n_1,m_0,m_1):
        return Matrix([row[n_0:n_1] for row in self.size[m_0:m_1]])
    
    
    
M1=Matrix([[2,4,6],[1,6,9],[3,2,1]])
    
print("Matrix:")
print(M1)

print("\nShape:")
print(M1.shape())

print("\nRow:")
print(M1.row(1))

print("\nCol:")
print(M1.column(1))

print("\nMatrix List:")
print(M1.to_list())

print("\nBlock:")
print(M1.block(1,3,0,2))
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n,m,c):
    return Matrix([[c]*m for _ in range(n)])

def zeros(n,m):
    return constant(n,m,0)

def ones(n,m):
    return constant(n,m,-1)

def eye(n):
    return Matrix([[1 if row==col else 0 for col in range(n)] for row in range(n)])


C=constant(2,9,4.3)
print(f"Constant:\n{C.to_list()}\n")

Z=zeros(1,2)
print(f"Zeros:\n{Z.to_list()}\n")

O=ones(2,3)
print(f"Ones:\n{O.to_list()}\n")

E=eye(3)
print(f"Identity:\n{E.to_list()}")
```
